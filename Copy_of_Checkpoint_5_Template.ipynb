```
# Question 1

# Create a function that will recieve a hole/cylinder's radius as an input and then returns the volume of that cylinder.

# For the purpose of this question you can use the value of 3.1415 as a constant for pi and 5 as a constant for height.
# Formula for the volume of a cylinder: pi*r^2*H  r = radius H = height

# Use this function to show the volume of a cylinder with a radius of 7 ft and then print out the result as a f string that says
# "Volume: {volume_variable} cubed feet." where the volume_variable is the value your function returns. Round your answer to the two decimal places.

pi = 3.1415
height_constant = 5

def cylinder_volume(radius):
  volume_variable = pi * height_constant * radius ** 2
  return volume_variable

cylinder_volume(7)
```
